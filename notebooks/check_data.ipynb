```python
import pandas as pd
import os

# CORRECT file names - matching your actual files
files = [
    'seoul-metro-2015.logs.csv',
    'seoul-metro-2016.logs.csv',
    'seoul-metro-2017.logs.csv',
    'seoul-metro-2018.logs.csv',
    'seoul-metro-2019.logs.csv',
    'seoul-metro-2020.logs.csv',
    'seoul-metro-2021.logs.csv',
]

print("="*65)
print(f"{'File':<35} {'Cols':>5} {'Size MB':>8}")
print("="*65)

for f in files:
    path = f'../data/{f}'
    if os.path.exists(path):
        df   = pd.read_csv(path, nrows=3)
        size = os.path.getsize(path) / (1024*1024)
        print(f"{f:<35} {len(df.columns):>5} {size:>7.1f}")
    else:
        print(f"{f:<35} FILE NOT FOUND")

print("="*65)
```

```python
# Check columns of each file
for f in files:
    path = f'../data/{f}'
    if os.path.exists(path):
        df = pd.read_csv(path, nrows=2)
        print(f"\n{f}")
        print(f"Columns: {df.columns.tolist()}")
        print(df.head(2))
```
